```python
import numpy as np
import pandas as pd
```

```python
#import nba player boxscores csv
player_boxscores = pd.read_csv('NBA box scores/2012-18_playerBoxScore.csv')
player_boxscores.head()
```

```python
#remove unneeded columns
player_boxscores = player_boxscores[['gmDate', 'gmTime','teamAbbr','teamRslt','playDispNm','playPTS','playAST','playMin','playTO','playSTL','playBLK','playPF','playFGA','playFGM','playFG%','play2PA','play2PM','play2P%','play3PA','play3PM','play3P%','playFTA','playFTM','playFT%','playORB','playDRB','playTRB', 'opptAbbr', 'opptRslt']]
player_boxscores.head()
```

```python
#import nba team boxscores csv
team_boxscores = pd.read_csv('NBA box scores/2012-18_teamBoxScore.csv')
team_boxscores.head()
```

```python
#remove unneeded columns and merge
team_boxscores = team_boxscores[['gmDate', 'gmTime','teamAbbr','teamPTS','teamAST']]
merged_boxscores = player_boxscores.merge(team_boxscores, on = ['gmDate','gmTime','teamAbbr'])
merged_boxscores.head()
```

```python
#make simplified df to work with
working_boxscores = merged_boxscores[['gmDate', 'teamAbbr','teamRslt','playDispNm','playPTS', 'teamPTS','teamAST']]
working_boxscores.head(15)
```

```python
#make needed columns: 'game_id', 'player_contribution'
#'season', 'assists/score', 'w_count', 'l_count', 'w/l_ratio'

#game_id, w/l
game_num = []
unique_id = 0
for row in range(len(working_boxscores['gmDate'])-1):
    if (working_boxscores.iloc[row+1, 0] != working_boxscores.iloc[row, 0]) or (working_boxscores.iloc[row+1, 1] != working_boxscores.iloc[row, 1]):
        game_num.append(unique_id)
        unique_id += 1
    else:
        game_num.append(unique_id)
game_num.append(unique_id)
working_boxscores = working_boxscores.assign(game_id=game_num)
working_boxscores.head()
```

```python
working_boxscores = working_boxscores[['game_id', 'teamAbbr','teamRslt','teamAST','playDispNm','playPTS','teamPTS']]
```

```python
#player_contribution
working_boxscores = working_boxscores.assign(
    player_contribution = working_boxscores['playPTS']/working_boxscores['teamPTS']*100)
working_boxscores.head()
```

```python
#season
year = 2012
for seasons in range(7):
    working_boxscores.loc[(working_boxscores.gmDate > f'{year}-10-0') & (working_boxscores.gmDate < f'{year+1}-5-0'), 'season'] = f'{year}'
    year += 1
working_boxscores
```

```python
#convert Win Loss to integers
working_boxscores.loc[working_boxscores.teamRslt == 'Loss', 'teamRslt'] = 0
working_boxscores.loc[working_boxscores.teamRslt == 'Win', 'teamRslt'] = 1
working_boxscores.head()
```

```python
#win_count and w_ratio
grouped_season_game_mean = working_boxscores.groupby(['season','teamAbbr', 'game_id']).mean().reset_index()
grouped_season_game_sum = grouped_season_game_mean.groupby(['season','teamAbbr']).sum().reset_index()
grouped_season_game_count = grouped_season_game_mean.groupby(['season','teamAbbr']).count().reset_index()
win_count = grouped_season_game_sum['teamRslt'].tolist()
w_ratio = (grouped_season_game_sum['teamRslt']/grouped_season_game_count['teamRslt']).tolist()
```

```python
#star_power = top1/avg(top5-top1)
star_calcs = []
for game in working_boxscores['game_id']:
    top_five = working_boxscores.loc[working_boxscores.game_id == game].nlargest(5, 'player_contribution')
    ['player_contribution'].tolist()
    star_calcs.append(top_five[0]/(np.mean(top_five[1:5])))
working_boxscores = working_boxscores.assign(star_power = star_calcs)
```

```python
#assists/score
grouped_season_game_mean = working_boxscores.groupby(['season','teamAbbr', 'game_id']).mean().reset_index()
grouped_season_game_mean = grouped_season_game_mean.assign(assists_score = grouped_season_game_mean['teamAST']/grouped_season_game_mean['teamPTS'])
```

```python
#add the columns
assists_score = grouped_season_game_mean.groupby(['season','teamAbbr']).mean()['assists_score'].tolist()
star_power = grouped_season_game_mean.groupby(['season','teamAbbr']).mean()['star_power'].tolist()
```

```python
final_df = working_boxscores.groupby(['season','teamAbbr']).mean().reset_index()[['season','teamAbbr']]
final_df['star_power'] = star_power
final_df['assists/score'] = assists_score
final_df['win_count'] = win_count
final_df['w/l_ratio'] = w_ratio
final_df.head(10)
```

```python
#there are 30 unique teams in the NBA vs 351 in the NCAA
final_df['teamAbbr'].nunique()
```

```python
#avg team_performance by year for top 10 teams
season_avg = []
for season in final_df['season'].unique():
    season_avg.append(final_df.loc[final_df.season == season].nlargest(3, 'win_count'))
top_ten_per_season = pd.concat(season_avg)
top_ten_per_season.head()
```

```python
top_ten_per_season.to_csv('top_ten_pct_nba.csv')
```

```python
#pivot that shit!
top_ten_by_season['stat']=str('star_power')
top_ten_star_power = top_ten_by_season.pivot(index = 'stat',columns = 'season',values = 'star_power_mean')
top_ten_by_season['stat']=str('assists/score')
top_ten_assists_score = top_ten_by_season.pivot(index = 'stat',columns = 'season',values = 'assists/score_mean')
top_ten_by_season['stat']=str('star_power_sem')
top_ten_star_sem = top_ten_by_season.pivot(index = 'stat',columns = 'season',values = 'star_power_sem')
top_ten_by_season['stat']=str('assists/score_sem')
top_ten_assists_sem = top_ten_by_season.pivot(index = 'stat',columns = 'season',values = 'assists/score_sem')
top_ten_pivot = pd.concat([top_ten_star_power, top_ten_assists_score,top_ten_star_sem,top_ten_assists_sem])
top_ten_pivot
```

```python
top_ten_pivot.to_csv('nba_top_ten_avg_by_season.csv')
```
