# League 2PT attempts vs 3PT attempts

```python
nba_17_18_x = pd.DataFrame(nba_17_18[['3P', '3PA', '2P', '2PA']])
nba_16_17_x = pd.DataFrame(nba_16_17[['3P', '3PA', '2P', '2PA']])
nba_15_16_x = pd.DataFrame(nba_15_16[['3P', '3PA', '2P', '2PA']])
nba_14_15_x = pd.DataFrame(nba_14_15[['3P', '3PA', '2P', '2PA']])
nba_13_14_x = pd.DataFrame(nba_13_14[['3P', '3PA', '2P', '2PA']])

nba_13_14_x['Year'] = '2014'
nba_14_15_x['Year'] = '2015'
nba_15_16_x['Year'] = '2016'
nba_16_17_x['Year'] = '2017'
nba_17_18_x['Year'] = '2018'

nba_13_14_x = nba_13_14_x.groupby(['Year'])
nba_13_14_x = (nba_13_14_x.sum()/30)
nba_14_15_x = nba_14_15_x.groupby(['Year'])
nba_14_15_x = nba_14_15_x.sum()/30
nba_15_16_x = nba_15_16_x.groupby(['Year'])
nba_15_16_x = nba_15_16_x.sum()/30
nba_16_17_x = nba_16_17_x.groupby(['Year'])
nba_16_17_x = nba_16_17_x.sum()/30
nba_17_18_x = nba_17_18_x.groupby(['Year'])
nba_17_18_x = nba_17_18_x.sum()/30
```

```python
combined_2and3tot = pd.merge(nba_13_14_x, nba_14_15_x, how='outer', on=['Year', '3P', '3PA', '2P', '2PA'])
combined_2and3tot = pd.merge(combined_2and3tot, nba_15_16_x, how='outer', on=['Year', '3P', '3PA', '2P', '2PA'])
combined_2and3tot = pd.merge(combined_2and3tot, nba_16_17_x, how='outer', on=['Year', '3P', '3PA', '2P', '2PA'])
combined_2and3tot = pd.merge(combined_2and3tot, nba_17_18_x, how='outer', on=['Year', '3P', '3PA', '2P', '2PA'])
combined_2and3tot
```

```python
plt.errorbar(combined_2and3tot.index, combined_2and3tot['3PA'],
             marker='o', linestyle='--')

plt.grid()
plt.legend(loc='best', fancybox=True)
plt.title("Change of 3PT Shots Over the Years")
plt.xlabel("Year")
plt.ylabel("3PT Shots Attempted (per game)")

# Save the Figure
plt.show()
```
