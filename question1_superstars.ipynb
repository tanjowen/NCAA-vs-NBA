```python
import pandas as pd
import numpy as np
```

```python
ncaa_df = pd.read_csv('NCAA data/cleaned_ncaa_data.csv')
ncaa_group = ncaa_df.groupby(['game_id','season','team_alias','full_name']).mean()
ncaa_df.head()
```

```python
#use dict to simplify ['game_id'] name
game_id_unique = ncaa_df['game_id'].unique()
game_id_replace = np.arange(0, len(game_id_unique), 1)
game_id_dict = dict(zip(game_id_unique, game_id_replace))
ncaa_df['game_id'] = ncaa_df.game_id.map(game_id_dict)
ncaa_df_index = ncaa_df.set_index('game_id')
ncaa_df.head()
```

```python
game_team_group_sum = ncaa_df.groupby(['game_id','team_alias']).sum()
game_team_group_sum.head(10)
```

```python
game_team_group_points_sum = game_team_group_sum[['points']]
game_team_group_points_sum = game_team_group_points_sum.reset_index()
game_team_group_points_sum.head()
```

```python
ncaa_df.team_alias[1]
```

```python
index_count = 0
ncaa_df['player_contribution'] = 0
for score in game_team_group_points_sum['points']:
    for index, row in ncaa_df_index.iterrows():
        if ncaa_df.loc[index, 'game_id'] == index_count and ncaa_df.loc[index, 'team_alias'] == ncaa_df.loc[(index+1), 'team_alias']:
            ncaa_df.player_contribution[index] = score
        else:
            ncaa_df.player_contribution[index] = score
            index_count += 1
            break
print(ncaa_df['player_contribution'])
#for index, row in ncaa_df_index.iterrows():
 #   for score in game_team_group_points_sum['points']:
  #      if ncaa_df_index[index, row] == index_count and ncaa_df[(index-1), 'team_alias'] == ncaa_df[(index-1), 'team_alias']:
   #         ncaa_df.player_contribution[row] = game_team_group_points_sum.points[score]
```
