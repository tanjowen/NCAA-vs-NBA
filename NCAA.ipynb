```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

ncaa_17_18 = pd.read_csv("NCAA data/ncaa data 17-18.csv")
ncaa_16_17 = pd.read_csv("NCAA data/ncaa data 16-17.csv")
ncaa_15_16 = pd.read_csv("NCAA data/ncaa data 15-16.csv")
ncaa_14_15 = pd.read_csv("NCAA data/ncaa data 14-15.csv")
ncaa_13_14 = pd.read_csv("NCAA data/ncaa data 13-14.csv")
```

```python
ncaa_13_14_x = ncaa_13_14.drop(ncaa_13_14.index[30:])
ncaa_14_15_x = ncaa_14_15.drop(ncaa_14_15.index[30:])
ncaa_15_16_x = ncaa_15_16.drop(ncaa_15_16.index[30:])
ncaa_16_17_x = ncaa_16_17.drop(ncaa_16_17.index[30:])
ncaa_17_18_x = ncaa_17_18.drop(ncaa_17_18.index[30:])
ncaa_13_14_x.head()
```

```python
ncaa_13_14_two = pd.DataFrame()
ncaa_14_15_two = pd.DataFrame()
ncaa_15_16_two = pd.DataFrame()
ncaa_16_17_two = pd.DataFrame()
ncaa_17_18_two = pd.DataFrame()
#find the amount of 2PA 
ncaa_13_14_2P_2PA = ncaa_13_14_x['FGA'] - ncaa_13_14_x['3PA']
ncaa_14_15_2P_2PA = ncaa_14_15_x['FGA'] - ncaa_14_15_x['3PA']
ncaa_15_16_2P_2PA = ncaa_15_16_x['FGA'] - ncaa_15_16_x['3PA']
ncaa_16_17_2P_2PA = ncaa_16_17_x['FGA'] - ncaa_16_17_x['3PA']
ncaa_17_18_2P_2PA = ncaa_17_18_x['FGA'] - ncaa_17_18_x['3PA']

#find the amount of 2P
ncaa_13_14_2P_ = ncaa_13_14_x['FG'] - ncaa_13_14_x['3P']
ncaa_14_15_2P_ = ncaa_14_15_x['FG'] - ncaa_14_15_x['3P']
ncaa_15_16_2P_ = ncaa_15_16_x['FG'] - ncaa_15_16_x['3P']
ncaa_16_17_2P_ = ncaa_16_17_x['FG'] - ncaa_16_17_x['3P']
ncaa_17_18_2P_ = ncaa_17_18_x['FG'] - ncaa_17_18_x['3P']

#calc the 2P per game
ncaa_13_14_two['2P'] = ncaa_13_14_2P_/ncaa_13_14_x['G']
ncaa_14_15_two['2P'] = ncaa_14_15_2P_/ncaa_14_15_x['G']
ncaa_15_16_two['2P'] = ncaa_15_16_2P_/ncaa_15_16_x['G']
ncaa_16_17_two['2P'] = ncaa_16_17_2P_/ncaa_16_17_x['G']
ncaa_17_18_two['2P'] = ncaa_17_18_2P_/ncaa_17_18_x['G']

#calc the 2PA per game
ncaa_13_14_two['2PA'] = ncaa_13_14_2P_2PA/ncaa_13_14_x['G']
ncaa_14_15_two['2PA'] = ncaa_14_15_2P_2PA/ncaa_14_15_x['G']
ncaa_15_16_two['2PA'] = ncaa_15_16_2P_2PA/ncaa_15_16_x['G']
ncaa_16_17_two['2PA'] = ncaa_16_17_2P_2PA/ncaa_16_17_x['G']
ncaa_17_18_two['2PA'] = ncaa_17_18_2P_2PA/ncaa_17_18_x['G']

#calc the 3P per game
ncaa_13_14_two['3P'] = ncaa_13_14_x['3P']/ncaa_13_14_x['G']
ncaa_14_15_two['3P'] = ncaa_14_15_x['3P']/ncaa_14_15_x['G']
ncaa_15_16_two['3P'] = ncaa_15_16_x['3P']/ncaa_15_16_x['G']
ncaa_16_17_two['3P'] = ncaa_16_17_x['3P']/ncaa_16_17_x['G']
ncaa_17_18_two['3P'] = ncaa_17_18_x['3P']/ncaa_17_18_x['G']

#calc the 3PA per game
ncaa_13_14_two['3PA'] = ncaa_13_14_x['3PA']/ncaa_13_14_x['G']
ncaa_14_15_two['3PA'] = ncaa_14_15_x['3PA']/ncaa_14_15_x['G']
ncaa_15_16_two['3PA'] = ncaa_15_16_x['3PA']/ncaa_15_16_x['G']
ncaa_16_17_two['3PA'] = ncaa_16_17_x['3PA']/ncaa_16_17_x['G']
ncaa_17_18_two['3PA'] = ncaa_17_18_x['3PA']/ncaa_17_18_x['G']

#add year to the dataframe
ncaa_13_14_two['Year'] = '2014'
ncaa_14_15_two['Year'] = '2015'
ncaa_15_16_two['Year'] = '2016'
ncaa_16_17_two['Year'] = '2017'
ncaa_17_18_two['Year'] = '2018'

#groupby year and find avg per game stats of top 30 teams
ncaa_13_14_two = ncaa_13_14_two.groupby(['Year'])
ncaa_13_14_two = (ncaa_13_14_two.sum()/30)
ncaa_14_15_two = ncaa_14_15_two.groupby(['Year'])
ncaa_14_15_two = ncaa_14_15_two.sum()/30
ncaa_15_16_two = ncaa_15_16_two.groupby(['Year'])
ncaa_15_16_two = ncaa_15_16_two.sum()/30
ncaa_16_17_two = ncaa_16_17_two.groupby(['Year'])
ncaa_16_17_two = ncaa_16_17_two.sum()/30
ncaa_17_18_two = ncaa_17_18_two.groupby(['Year'])
ncaa_17_18_two = ncaa_17_18_two.sum()/30

ncaa_13_14_two
```

```python
combined_ncaa_2and3tot = pd.merge(ncaa_13_14_two, ncaa_14_15_two, how='outer', on=['Year', '3P', '3PA', '2P', '2PA'])
combined_ncaa_2and3tot = pd.merge(combined_ncaa_2and3tot, ncaa_15_16_two, how='outer', on=['Year', '3P', '3PA', '2P', '2PA'])
combined_ncaa_2and3tot = pd.merge(combined_ncaa_2and3tot, ncaa_16_17_two, how='outer', on=['Year', '3P', '3PA', '2P', '2PA'])
combined_ncaa_2and3tot = pd.merge(combined_ncaa_2and3tot, ncaa_17_18_two, how='outer', on=['Year', '3P', '3PA', '2P', '2PA'])
combined_ncaa_2and3tot
```

```python
plt.errorbar(combined_ncaa_2and3tot.index, combined_ncaa_2and3tot['3PA'],
             marker='o', linestyle='--')

plt.grid()
plt.legend(loc='best', fancybox=True)
plt.title("Change of 3PT Shots Over the Years")
plt.xlabel("Year")
plt.ylabel("3PT Shots Attempted (per game)")

# Save the Figure
plt.show()
```
